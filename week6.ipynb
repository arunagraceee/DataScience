```python
import pandas as pd
df=pd.read_csv("Iris.csv")
print(df.head())
mean_values=df.mean(numeric_only=True)
print(mean_values)
median_values=df.median(numeric_only=True)
print(median_values)
mode_values=df.mode().iloc[0]
print(mode_values)
```

```python
#measures of dispersion
import pandas as pd
df=pd.read_csv("Iris.csv")
print(df.head())
range_values=df.max(numeric_only=True)-df.min(numeric_only=True)
print(range_values)
variance_values=df.var(numeric_only=True)
print(variance_values)
std_values=df.std(numeric_only=True)
print(std_values)
Q1=df.quantile(0.25,numeric_only=True)
Q3=df.quantile(0.75,numeric_only=True)
IQR=Q3-Q1
print(IQR)
```

```python
#shape of distibution
import pandas as pd
df=pd.read_csv("Iris.csv")
print(df.head())
skewness_values=df.skew(numeric_only=True)
print(skewness_values)
kurtosis_values=df.kurtosis(numeric_only=True)
print(kurtosis_values)
```

```python
# Correlation and Association

import pandas as pd
from scipy.stats import pearsonr

iris = pd.read_csv("Iris.csv")

# Pearson Correlation
corr, p_value = pearsonr(iris['SepalLengthCm'], iris['PetalLengthCm'])

print(f"Pearson correlation coefficient: {corr}")
print(f"P-value: {p_value}")

# Correlation Matrix
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_iris.corr(method='pearson')

print("\nCorrelation Matrix:")
print(corr_matrix)
```
